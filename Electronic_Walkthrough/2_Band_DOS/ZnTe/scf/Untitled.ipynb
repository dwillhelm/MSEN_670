```python
import pymatgen as mg
from pymatgen.electronic_structure.plotter import * 
from pymatgen.io.vasp.outputs import * 
```

```python
vr = Vasprun('vasprun.xml') 
bs = vr.get_band_structure()
dos = vr.complete_dos
```

```python
myfig = BSPlotter(bs)
myfig.get_plot()
```

```python
myfig.save_plot(filename='Cu_BS',img_format='png')
```

```python
bsdos = BSDOSPlotter(vb_energy_range=10,cb_energy_range=10,egrid_interval=2)
myfig = bsdos.get_plot(bs,dos)
```
